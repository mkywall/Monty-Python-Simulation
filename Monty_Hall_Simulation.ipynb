```python
#1. Create an list of lists with A,B,C repeated X times.
all=[('A','B','C')]*10000
all
```

```python
#2. From this list, randomly pick one door which contains the car and create a list.
import random
car_door = [random.choice(each_trial) for each_trial in all] 
car_door
```

```python
#3. Using set operations generate a list of lists of doors which contain goats.
goat_doors=list([set(each_all)-set(each_car_door) for each_all in all for each_car_door in car_door])
goat_doors
```

```python
#4. User selection will be a list of randomly selected doors from list created in step 1.
user_pick = [random.choice(each_trial) for each_trial in all] 
user_pick
```

```python
#5. Using set operations generate a list of lists of doors which were not selected by the User.
not_user_pick=list([set(each_all)-set(each_user_pick) for each_all in all for each_user_pick in user_pick])
not_user_pick
```

```python
#6. Using set operations (intersection) select (one) door which is common in list generated in step 5 and 3.
# This is the door opened by the presenter.
presenter_open=[]
for i in range(10000):
    x=not_user_pick[i]&goat_doors[i]
    if len(x)>1:
        x=random.choice(list(x))
    presenter_open.append(set(x))
presenter_open
```

```python
#7.1. Perform switch operation by performing un-intersect of step 5 and 6. 
#   Compare this list to list created in step 2 and get match %. This is the 
#   winning percentage.

switch_door=[]
for i in range(10000):
    x=not_user_pick[i]^presenter_open[i]
    switch_door.append(x)
#switch_door
win_switch=0
for i in range(10000):
    if set(switch_door[i])==set(car_door[i]):
        win_switch=win_switch+1
win_switch
```

```python
#7.2. Perform no operation and compare original user selection with list created 
#    in step 2 and get match %. This is the winning percentage.
win_original=0
for i in range(10000):
    if set(user_pick[i])==set(car_door[i]):
        win_original=win_original+1
win_original
```

```python
#8. Compare the winning % in both the cases. Make observation on switch vs no switch.
print("The winning percentage if a contestant switches door is: ", 100*win_switch/10000)
print("The winning percentage if a contestant does not switch door is: ", 100*win_original/10000)
```
